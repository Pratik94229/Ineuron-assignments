```
Q1. What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Ans. Matplotlib is a popular data visualization library in Python that provides a wide range of functionalities for creating high-quality plots, graphs, and charts. It is widely used for visualizing and analyzing data in various fields such as data science, machine learning, statistics, and more.

Here are five plots that can be created using the Pyplot module of Matplotlib:

1. Line Plot: A line plot is used to represent data points connected by straight lines. It is useful for visualizing trends and patterns over a continuous variable or time.

2. Scatter Plot: A scatter plot displays individual data points as markers on a two-dimensional plane. It is helpful for examining the relationship between two variables and identifying any patterns or clusters.

3. Bar Plot: A bar plot (or bar chart) uses rectangular bars to represent categorical data. It is commonly used to compare and visualize the distribution or frequency of different categories.

4. Histogram: A histogram displays the distribution of a continuous variable by dividing it into bins and representing the frequency or count of data points falling into each bin. It helps visualize the shape, central tendency, and spread of the data.

5. Pie Chart: A pie chart represents data as a circular graph divided into slices, where each slice represents a proportion or percentage of the whole. It is suitable for displaying categorical data and comparing different categories based on their relative sizes.

These are just a few examples of the plots that can be created using the Pyplot module of Matplotlib. Matplotlib provides many other types of plots, such as box plots, violin plots, heatmaps, and more, allowing users to choose the appropriate visualization technique based on their data and analysis goals.
```

```python
import numpy as np

x = 3 + np.random.normal(0, 2, 50)
```

```
Note: Also add title, xlabel, and ylabel to the plot.

Ans. A scatter plot is a type of plot used to display the relationship between two numerical variables. It represents individual data points as markers on a two-dimensional plane, with one variable plotted along the x-axis and the other variable plotted along the y-axis.

The scatter plot is particularly useful for examining the correlation or association between the two variables and identifying any patterns or trends in the data. Each data point on the plot represents a unique combination of values for the two variables, allowing for the visualization of the distribution and clustering of the data.

Scatter plots can provide insights into several aspects of the relationship between variables, including:

1. Correlation: By observing the general trend of the data points on the scatter plot, you can determine whether there is a positive, negative, or no correlation between the two variables. A positive correlation means that as one variable increases, the other tends to increase as well, while a negative correlation indicates an inverse relationship.

2. Outliers: Scatter plots can reveal outliers, which are data points that deviate significantly from the overall pattern. Outliers may indicate errors in data collection or represent unique cases that require further investigation.

3. Clustering: Patterns or clusters of data points on the scatter plot can suggest the presence of groups or subgroups within the data. This can be helpful for identifying distinct patterns or segments in the dataset.

4. Distribution: Scatter plots provide a visual representation of the distribution of data points. The concentration and spread of the points can give insights into the variability and density of the data.

Scatter plots can be enhanced with additional features such as color-coding, size variation, or regression lines to provide further information or highlight specific characteristics in the data.
```

````python
import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

import matplotlib.pyplot as plt

plt.scatter(x,y)

plt.show("svg")
````

```
Q3. Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:
```
```python
import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```
```
Ans. The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to divide the figure into multiple smaller plots, each with its own axes and content. The subplot() function is helpful in several scenarios:

1. Creating Multiple Plots: The subplot() function provides a convenient way to create multiple plots within a single figure. By specifying the number of rows, number of columns, and the position of each subplot in the grid, you can easily arrange and visualize multiple plots side by side.

2. Comparing Data: When you have multiple datasets or variations of data that you want to compare, using subplots can be beneficial. Each subplot can represent a different aspect of the data, allowing for easier visual comparison and analysis.

3. Customizing Layout: The subplot() function allows you to have control over the layout of the subplots. You can specify the size, position, and spacing of each subplot, enabling you to create complex arrangements of plots according to your specific requirements.

4. Grouping Related Plots: Subplots are useful when you want to group related plots together. For example, you may want to display multiple views or different visualizations of the same data in a single figure. Subplots provide a concise way to organize and present such information.

5. Storytelling and Presentation: Subplots are often used in presentations or reports to present a series of related visualizations or to show the progression of data over time. They can be used to tell a data-driven story or present different aspects of an analysis in a structured manner.

By utilizing the subplot() function, you can create more complex and informative figures by combining multiple plots into a single layout. This allows for better visualization, comparison, and presentation of data and enhances the overall understanding of the information being conveyed.
```

````python
import numpy as np
import matplotlib.pyplot as plt

#Line 1

x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])

#Line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

#Line 3
x3 = np.array([0, 1, 2, 3, 4, 5])  
y3 = np.array([10, 20, 30, 40, 50, 60])

#Line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create figure
fig = plt.figure()

# Plot on each subplot
plt.plot(x1, y1, label = 'Line 1')
plt.plot(x2, y2, label = 'Line 2')
plt.plot(x3, y3, label = 'Line 3')
plt.plot(x4, y4, label = 'Line 4')
plt.legend()
# Set title
plt.title("Multiple Line Plot")

# Display the figure
plt.show("svg")
````

```
Q4. What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
```
```python
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
```
```
Ans. A bar plot, also known as a bar chart, is a visualization tool used to represent categorical data with rectangular bars. It is used to compare and display the distribution or frequency of different categories or groups.

The primary purpose of a bar plot is to provide a visual summary of categorical data and highlight any disparities or patterns among the categories. Here are some key characteristics and uses of bar plots:

1. Categorical Comparison: Bar plots are effective in comparing the values of different categories or groups. The length or height of each bar represents the quantity or frequency of the category it represents, allowing for easy visual comparison.

2. Data Distribution: Bar plots provide insights into the distribution of data across categories. They help identify the categories with the highest or lowest values, allowing for quick identification of trends or outliers.

3. Data Exploration: Bar plots are useful for exploring data and identifying patterns or relationships between categories. They can reveal patterns such as trends, seasonality, or variations in data across different groups.

4. Data Comparison: Bar plots can be used to compare data across multiple variables or subgroups within categories. By using different colors or grouping bars together, it becomes possible to compare values across different dimensions or conditions.

5. Communication and Presentation: Bar plots are commonly used for data visualization in presentations, reports, and publications. They provide a clear and intuitive representation of categorical data, making it easier for the audience to understand and interpret the information.

6. Decision Making: Bar plots aid decision-making processes by providing a visual summary of categorical data. They help stakeholders identify patterns, make comparisons, and draw insights that can support informed decision-making.

Bar plots can be customized to enhance their visual appeal and convey additional information. For example, labels, titles, and color-coding can be added to provide context and improve interpretability. Bar plots are versatile and widely used in various domains, including business, marketing, social sciences, and data analysis, to present and analyze categorical data effectively.
```

````python
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
````

````python
# Bar Plot

plt.bar(company, profit)
plt.xlabel("Companies")
plt.ylabel("Profit")
plt.title("Company and Their Profits")
plt.show("svg")
````

````python
# Horizontal Bar Plot

plt.barh(company, profit)
plt.ylabel("Companies")
plt.xlabel("Profit")
plt.title("Company and Their Profits")
plt.show("svg")
````

```
Q5. What is a box plot? Why is it used? Using the following data plot a box plot.
```
```python
box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)
```
```
Ans. A box plot, also known as a box-and-whisker plot, is a statistical visualization tool that provides a summary of the distribution of numerical data. It displays key statistical measures and identifies potential outliers in a dataset. Box plots are commonly used for the following purposes:

1. Data Summary: A box plot provides a concise summary of the distribution of a dataset. It displays key statistical measures such as the median, quartiles (25th and 75th percentiles), and the range of the data. The central box in the plot represents the interquartile range (IQR), which contains the middle 50% of the data. The line inside the box represents the median value.

2. Outlier Detection: Box plots are effective in identifying potential outliers in a dataset. Outliers are data points that significantly deviate from the rest of the data and may indicate anomalies or errors. Box plots help to visually identify values that fall beyond the whiskers (lines extending from the box), which are typically set to 1.5 times the IQR.

3. Comparison of Groups: Box plots are useful for comparing the distributions of multiple groups or categories. By placing multiple box plots side by side, it becomes easy to compare the central tendency, spread, and skewness of the data across different groups. This allows for the detection of variations, patterns, or differences between groups.

4. Skewness and Symmetry: Box plots provide insights into the skewness and symmetry of the data. If the box is symmetrical, it indicates that the data is evenly distributed around the median. Skewness can be observed when one whisker is longer than the other, suggesting that the data is positively or negatively skewed.

5. Data Visualization and Communication: Box plots are a popular choice for visualizing and communicating data in various fields such as statistics, data analysis, and research. They provide a clear and standardized representation of the distribution and key statistics, enabling easy understanding and comparison of datasets.

Box plots are versatile and can be customized to include additional information such as labels, titles, and annotations to provide context and enhance interpretability. They are particularly useful when dealing with large datasets, comparing groups, or analyzing distributions, making them a valuable tool for data exploration and analysis.
```

````python
fig = plt.figure()

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

ax1= fig.add_subplot(1,2,1)
ax1.boxplot(box1)
plt.title("Box Plot 1")

ax2= fig.add_subplot(1,2,2)
ax2.boxplot(box2)
plt.title("Box Plot 2")
plt.show("svg")
````
